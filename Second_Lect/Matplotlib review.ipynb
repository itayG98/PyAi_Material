# Matplotlib

## Plotting categorial vars 

```python
names = ['Apple', 'Orange', 'Banana']
values = [10, 7, 12.5]

fig , axes = plt.subplots(nrows=1 , ncols =3 , figsize=(9,3))

axes[0].bar(names, values)
axes[1].scatter(names, values)
axes[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

## Bar plot

```python
plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots( figsize=(3,2))

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

## x y simple example

# Marker and Colors

```python
x = np.arange(0,300,50)
fig , axis = plt.subplots(figsize=(7,3))
axis.grid(True, linestyle='-.')
axis.tick_params(labelcolor='blue', labelsize='large', width=2)

#markers
axis.plot(x,x,marker='+')
axis.plot(x,x*2,marker='o')
axis.plot(x,x*3,marker='*')
axis.plot(x,x*4,marker='s')

#colors
axis.plot(x,x*5,color='red')
axis.plot(x,x*6,color='purple')
axis.plot(x,x*7,color='green')

#line tpye
axis.plot(x,x*8,ls=':')
axis.plot(x,x*9,ls='-.')
axis.plot(x,x*10,ls='--')

#mix 
axis.plot(x,x*11,ls='-.' , color='orange',marker='s')
axis.plot(x,x*12,ls='--' ,color='c' ,marker='o')
```

## Load Excel file

```python
python_course_xlsx = pd.read_excel("python_course.xlsx",header=1)
```

```python
python_course_fig = plt.figure(figsize=(5,2))
ax = python_course_fig.add_axes([0.1,0.1,1,1])

ax.plot(python_course_xlsx['Year'],python_course_xlsx['Grade'])

ax.set_xlabel('Years')
ax.set_ylabel('Grade')
ax.set_title('Python Grades over time')
```

## Another plot example

```python
fig , axes = plt.subplots(figsize=(7,2))
axes.plot(x,y ,label ='Regular' ,color='r')
axes.plot(x,y+250, label ='y+5',color='g')
axes.plot(x,y*3, label ='y3' ,color='orange')
axes.legend(loc=9)
```

```python
fig = plt.figure()
fig.set_size_inches(5, 3)
axes_1 = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.1,0.1,0.2,0.2])
axes_1.plot(x,wired_func ,'r--' ,lw = 0.5,label = 'Wired func')
axes_2.plot(x,y , 'g--')
axes_1.legend()
```

## Both plots on the same figure
