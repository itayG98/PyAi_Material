## Create a figure from scratch X Y Plot

```python
python_course_xlsx = pd.read_excel("python_course.xlsx",header=1)
```

```python
python_course_fig = plt.figure()
ax = python_course_fig.add_axes([0.1,0.1,1,1])
ax.plot(python_course_xlsx['Year'],python_course_xlsx['Grade'])
ax.set_xlabel('Years')
ax.set_ylabel('Grade')
ax.set_title('Python Grades over time')
```

## Making subplots

```python
sub_plot_fig ,ax = plt.subplots(nrows=1 , ncols =2)
plt.tight_layout()
ax[0].plot(python_course_xlsx['Year'],python_course_xlsx['Grade'],'g*-')
ax[1].plot(python_course_xlsx['Year'],python_course_xlsx['Attendance'],'r+-')
#ax[2].plot(python_course_xlsx['Year'],python_course_xlsx['Attendance'],'c--')
```

```python
x = np.linspace(0,10,100)
y= x**2
wired_func = np.sin(y)*y 
```

```python
colors = ['b','g' , 'r' ,'c']
```

## Another plot example

```python
fig , axes = plt.subplots(figsize=(7,4))
axes.plot(x,y ,label ='Regular')
axes.plot(x,y+250, label ='y+5')
axes.plot(x,y*3, label ='y3')
axes.legend(loc=9)
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.1,0.1,0.1,0.1])
axes_1.plot(x,wired_func ,'r--' ,lw = 0.5,label = 'Wired func')
axes_2.plot(x,y , 'g--')
axes_1.legend()
```

# Marker and Colors

```python
x = np.arange(0,300,50)
fig , axis = plt.subplots(figsize=(10,5))
axis.grid(True, linestyle='-.')
axis.tick_params(labelcolor='blue', labelsize='large', width=2)

#markers
axis.plot(x,x,marker='+')
axis.plot(x,x*2,marker='o')
axis.plot(x,x*3,marker='*')
axis.plot(x,x*4,marker='s')

#colors
axis.plot(x,x*5,color='red')
axis.plot(x,x*6,color='purple')
axis.plot(x,x*7,color='gray')

#line tpye
axis.plot(x,x*8,ls=':')
axis.plot(x,x*9,ls='-.')
axis.plot(x,x*10,ls='--')

#mix 
axis.plot(x,x*11,ls='-.' , color='blue',marker='s')
axis.plot(x,x*12,ls='--' ,color='grey' ,marker='o')
```
